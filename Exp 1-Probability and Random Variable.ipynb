```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#a
X=np.random.normal(27,1,10**6)
Y=np.random.normal(27,1,10**6)
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.hist(X)
ax2.hist(Y)
```

```python
a1,b1=np.histogram(X, bins=100, density=True)
a2,b2=np.histogram(Y, bins=100, density=True)
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.plot(b1[:-1],a1)
ax2.plot(b2[:-1],a2)
```

```python
a=np.arange(20,35,0.1)
b=(1/(1*(np.sqrt(2*np.pi))))*(np.exp((-1/2)*(((a-27)/1)**2)))
plt.plot(a,b)
```

```python
#b
c1=np.cumsum(a1)
c2=np.cumsum(a2)
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.plot(b1[:-1],c1)
ax2.plot(b2[:-1],c2)
```

```python
X_new=np.round(X)
Y_new=np.round(Y)
a1n,b1n=np.histogram(X_new, bins=25, density=True)
a2n,b2n=np.histogram(Y_new, bins=25, density=True)
fig, (ax1,ax2)=plt.subplots(2,1)
ax1.bar(b1n[:-1],a1n)
ax2.bar(b2n[:-1],a2n)
```

```python
#c
Z=X+Y
plt.hist(Z)
print(np.mean(Z))
print(np.var(Z))
```

```python
a3,b3=np.histogram(Z, bins=100, density=True)
plt.plot(b3[:-1],a3)
```

```python
c3=np.cumsum(a3)
plt.plot(b3[:-1],c3)
```
